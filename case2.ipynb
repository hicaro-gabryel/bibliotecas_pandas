# Importação das bases e Tratamento dos DatasFrames

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
#verificando se o DF contém dados nulos
vendas.info()
```

```python
#analisando a quantidade de vendas por canal de veda
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
#substituindo o nome APP por Aplicativo
vendas['idcanalvenda'] = vendas ['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
#analisando a quantidade de vendas por departamento
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#substituindo o espaçõe entre as palavras por '_'
vendas['Nome_Departamento'] = vendas ['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#Analisando as vendas por estado
vendas.groupby('estado').idcompra.nunique()
```

```python
#filtrando os nulos
vendas[vendas['estado'].isnull()]
```

```python
#Eliminando os dados nulos
#(considerando que uma das premissas e a de não excluir as linhas que estão vazias, esta celula não deverá ser executada)
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()

```

```python
#Filtrando os dados nulos na coluna estado e substituido por MS
vendas['estado'].fillna('MS',inplace = True)
vendas.info()
```

```python
#Filtrando os dados nulos na coluna Preço e substituido pela media da coluna Preço
media_preco = vendas['Preço'].mean()
vendas ['Preço'].fillna(media_preco, inplace = True)
```

```python
#verificando se os dados ja foram completamente tratados
vendas.info()
```

```python
#analise inicial a respeto dos dados do DF
vendas[['Preço','Preço_com_frete']].describe()
```

```python
# criando condição para filtrar apenas os dados em que o preço final de um produto não excede o preço com frete.
preco_errado = vendas.query('Preço > Preço_com_frete')
vendas_correta = vendas.query('Preço < Preço_com_frete ')
vendas_correta.head()
```

```python
vendas_correta.query('Nome_Departamento == "Esporte_e_Lazer" and estado == "SP"').describe()
```

```python
#Departamentos mais vendidos
departamentos_vendidos = vendas_correta.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending =False).reset_index()
departamentos_vendidos
```

```python
#Média de preço com frete por Nome de Departamento
media_vendidos = round(vendas_correta.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending =False).reset_index(),2)
media_vendidos
```

```python
#Quantidade de vendas por Mês
vendas_mes = vendas_correta.groupby('Data').idcompra.nunique().sort_values(ascending = False)
vendas_mes.head()
```

```python
vendas_correta['mes'] = vendas_correta['Data'].dt.month
```

```python
vendas_mes2= vendas_correta.groupby('mes').idcompra.nunique().sort_values(ascending = False)
vendas_mes2
```

```python
#importando 2º Base
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
#analisando o 2ºDF
cliente[['idade','renda']].describe()
```

```python
#verificando se não tem dados nulos e o tipo das colunas
cliente.info()
```

```python
#alterando o tipo da coluna renda de int para float
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
 #Unindo os 2 DF's utilizando o merge através da coluna cliente_log presente em ambas as tabelas.
 vendas_cliente = vendas_correta.merge(cliente, left_on = 'cliente_Log', right_on = 'cliente_Log')
 vendas_cliente.head()
```

```python
#media de renda por canal de venda
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
#media de idade por bandeira
agg_band_idade =round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_band_idade
```

#Plotando os graficos utilizando as bibliotecas *matplotlib* e *plotly*

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'],agg_idcanal_renda['renda'], color ='green')

plt.ylabel('Renda Média ')

plt.title('Média de Renda por Canal de Venda', loc='center',fontsize=18,  fontweight='bold', color='black')
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_band_idade['bandeira'],agg_band_idade['idade'], color ='blue')

plt.ylabel('Média de Idade')

plt.title('Média de Idade por Bandeira', loc='center',fontsize=18,  fontweight='bold', color='black')
```
